```python
# elimina clumnas que tienen SOLO valores nulos
for columna in df1.columns:
    if df1[columna].isnull().all():
        df1.drop(columna, axis=1, inplace=True)
```

```python
#df categoricas
categoricas = df.select_dtypes(include=['object', 'category'])
#df numericas
df_numeric = df.select_dtypes(include=['number'])
```

#Outliers con rango intercuantil

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df_numeric

Q1=y.quantile(0.25) #Q1
Q3=y.quantile(0.75) #Q3
iqr= Q3 - Q1

Limite_Superior_iqr= Q3 + 1.5*iqr
Limite_Inferior_iqr= Q1 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#eliminamos outliers
#Obtenemos datos solo con los que estan dentro de los limites
dentro= df_numeric[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
dentro
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=dentro.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
clean =dentro.copy()
clean= clean.fillna(round(dentro.mean(),1))
clean
```

```python
valores_nulos=clean.isnull().sum()
valores_nulos
```

```python
for variable in primer:
    print("=" * 50)
    print(f"Ejecutando para la variable: {variable}")
    print("=" * 50)

    # Crear una tabla de frecuencia
    table1 = freq_tbl(df[variable])

    # Imprimir la tabla de frecuencia
    print(table1)

    # Filtrar la tabla de frecuencia
    filtro = table1[table1['frequency'] < 100]

    # Ajustar el índice del DataFrame filtrado
    filtro_index = filtro.set_index(variable)

    # Realizar un gráfico de barras del DataFrame filtrado
    filtro_index.plot(kind='bar', width=1, figsize=(10, 4))
    plt.title(variable)
    plt.xlabel('Nivel de acuerdo')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)

    # Realizar un gráfico de dispersión del DataFrame filtrado
    filtro_index.plot("frequency", "cumulative_perc", kind="scatter")

    # Realizar un gráfico de área del DataFrame filtrado
    filtro_index.plot(kind='area', figsize=(10, 4), alpha=0.8)

    # Mostrar los gráficos
    plt.show()
```
